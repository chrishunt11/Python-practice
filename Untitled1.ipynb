```python
from nba_api.stats.endpoints import leaguegamefinder

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

pd.set_option('display.max_columns', None)
```

```python
# Step 1: Fetch Data from NBA API
# Fetch NBA game data for a specific season
gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable='2024-25')  # Replace with desired season
games = gamefinder.get_data_frames()[0]  # Get the first DataFrame from the response

# Display the first few rows to confirm data structure
print("Initial data:")
print(games.head())
```

```python
# Step 2: Check the structure of the dataset
print("\nColumns in the dataset:")
print(games.columns)
```

```python
# Rename columns for better readability
games.rename(columns={
    'TEAM_ID': 'Team_ID',
    'PTS': 'Points',
    'WL': 'Win_Loss'
}, inplace=True)
```

```python
# Step 4: Feature Engineering
# Extract "home" and "away" teams from the 'MATCHUP' column
games['Home_Team'] = games['MATCHUP'].apply(lambda x: x.split(' vs. ')[0] if 'vs.' in x else None)
games['Away_Team'] = games['MATCHUP'].apply(lambda x: x.split(' vs. ')[1] if 'vs.' in x else None)

# Convert 'WL' to a binary win indicator
games['Home_Team_Wins'] = games['Win_Loss'].apply(lambda x: 1 if x == 'W' else 0)

print("\nData with engineered features:")
print(games.head())
```

```python
# Create an empty list to store all team stats
all_team_stats = []

# Iterate through each team
for team in unique_teams:
    team_games = nba_teams[nba_teams['TEAM_NAME'] == team]
    win_loss_count = team_games.groupby('Win_Loss').size().reset_index(name='Game_Count')
    avg_stat = team_games.groupby('Win_Loss')[['Points', 'FG3M', 'FG3A', 'FG3_PCT', 'REB', 'STL', 'TOV']].mean()
    combined_stats = pd.merge(win_loss_count, avg_stat, on='Win_Loss').round(2)
    combined_stats['TEAM_NAME'] = team  # Add team name for identification
    combined_stats = combined_stats.sort_values(by='Win_Loss', ascending=False)  # Wins on top
    all_team_stats.append(combined_stats)

# Concatenate all team stats into a single DataFrame
all_team_stats_df = pd.concat(all_team_stats)

# Sort by the most wins (Win_Loss == 'W') to least wins
sorted_stats = all_team_stats_df[all_team_stats_df['Win_Loss'] == 'W'].sort_values(by='Game_Count', ascending=False)

# Append the losses for the sorted teams
final_sorted_stats = pd.concat(
    [
        sorted_stats,  # Wins sorted by most to least
        all_team_stats_df[all_team_stats_df['Win_Loss'] == 'L'].set_index('TEAM_NAME').loc[sorted_stats['TEAM_NAME']].reset_index()
    ]
)

# Display the final sorted stats
print(final_sorted_stats)
```

```python
# Combine wins and losses into a single table
standings = all_team_stats_df.pivot_table(
    index='TEAM_NAME',
    columns='Win_Loss',
    values=['Game_Count', 'Points', 'FG3M', 'FG3A', 'FG3_PCT', 'REB', 'STL', 'TOV'],
    aggfunc='first'  # Ensures unique stats for each Win_Loss category
).fillna(0)

# Flatten MultiIndex columns for better readability
standings.columns = [f"{col[0]}_{col[1]}" for col in standings.columns]

# Add total games column (Wins + Losses)
standings['Total_Games'] = standings['Game_Count_W'] + standings['Game_Count_L']

# Sort by the number of wins
standings = standings.sort_values(by='Game_Count_W', ascending=False)

# Rename columns for clarity
standings.rename(columns={
    'Game_Count_W': 'Wins',
    'Game_Count_L': 'Losses',
    'Points_W': 'Avg Points (Wins)',
    'Points_L': 'Avg Points (Losses)',
    'FG3M_W': '3PT Made (Wins)',
    'FG3M_L': '3PT Made (Losses)',
    'FG3A_W': '3PT Attempts (Wins)',
    'FG3A_L': '3PT Attempts (Losses)',
    'FG3_PCT_W': '3PT% (Wins)',
    'FG3_PCT_L': '3PT% (Losses)',
    'REB_W': 'Rebounds (Wins)',
    'REB_L': 'Rebounds (Losses)',
    'STL_W': 'Steals (Wins)',
    'STL_L': 'Steals (Losses)',
    'TOV_W': 'Turnovers (Wins)',
    'TOV_L': 'Turnovers (Losses)'
}, inplace=True)

# Display the standings
print(standings)

# Save to CSV
standings.to_csv('nba_standings_with_stats.csv', index=True)
print("Standings with stats saved to 'nba_standings_with_stats.csv'")
```
